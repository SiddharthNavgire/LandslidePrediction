```python
import pandas as pd

df = pd.read_csv("cleaned.csv")
df = df.drop('ID', axis=1)
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
df.columns
# Calculate mean for each column
mean_stats = df.mean()

# Calculate median for each column
median_stats = df.median()

# Calculate mode for each column
mode_stats = df.mode().iloc[0]  # Take the first row as mode

# Create a DataFrame with the calculated statistics
summary_stats = pd.DataFrame({'Mean': mean_stats, 'Median': median_stats, 'Mode': mode_stats})

# Format the DataFrame to display values without scientific notation
pd.set_option('display.float_format', '{:.4f}'.format)

# Print the result
print(summary_stats)
```
